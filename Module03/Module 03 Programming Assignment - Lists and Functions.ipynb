## Exercise 9.2
> Define a generator function called *get_odds()* that returns the odd number from range(10). Use a **for** loop to find and print and print the third value returned.

```python
def get_odds():
    """Designed to create a generator object which will yield odd numbers."""
    for num in range(10):
        if num % 2 == 1:
            yield num

iter_obj = get_odds()
count: int = 0

for iterable in iter_obj:
    print(iterable)
    count += 1
    if count == 3:
        third_value: int = iterable
print(f'The third value in our iterable object is {third_value}.')
```
