## Sort 0s, 1s, and 2s
> Given an array **arr[]** containing only **0s, 1s, and 2s**. Sort the array in ascending order. **Note:** You need to solve this problem wihtout utilizing the built-in sort function.


The code I provide below was my attempt at solving this problem without relying on the qsort function. The difficulty is in ensuring that is is truly a one-pass algorithm. This program works if the user is aware of the constraints. In this case, I know that I am only dealing with 0s, 1s, and 2s. There may be other methods for using a similar function on larger sets with unkown constraints, but research has shown that using list comprehension comparing to an item in the list is the more effecient method of sorting a list of unknown constraints. 

```python
unsorted: list = [0, 2, 1, 2, 0, 0, 1, 0, 2, 1, 2, 2, 0, 0, 1]
high: list = []
mid: list = []
low: list = []

class mysort:
    def __init__(self, x):
        self.high_range = 2
        self.mid_range = 1
        self.low_range = 0
        self.x = x

 
    def compare(self):
        if self.x <= self.low_range:
            low.append(self.x)
        elif self.x >= self.high_range:
            high.append(self.x)
        else:
            mid.append(self.x) 

for x in unsorted:
    instance = mysort(x)
    instance.compare()

print(low + mid + high)
```

```python
unsorted2: list = [0, 2, 1, 2, 0, 0, 1, 0, 2, 1, 2, 2, 0, 0, 1]
def sort_by_comprehension(arr):
    """ This function will sort the list provided by first identifying a reference point called pivot and comparing this value through
     iteration and place the compared values in separate temporary lists which will be concatenated at the end."""
    
       
    if len(arr) <= 1:
        return arr
    
    pivot: int = arr[len(arr)//2]
 
    
    left: list = [i for i in arr if i < pivot]
    center: list = [i for i in arr if i == pivot]
    right: list = [i for i in arr if i > pivot]

    return sort_by_comprehension(left) + center + sort_by_comprehension(right)

print(sort_by_comprehension(unsorted2))

```

## Binary Search
> Given a sorted array arr[] and an integer k, find the position(0-based indexing) at which k is present in the array using binary search. If k doesn't exist in arr[] return -1. 

**Note:** If multiple occurrences are there, please return the smallest index.

```python
import random

list_size: int = random.randint(10, 50)
lst: list = sorted([(random.randint(5, 95)) for num in range(list_size)]) 
k: int = random.choice(range(1,100))
list_length: int = len(lst)
i: int = lst[len(lst)//2]
remember_index: int = 0
first_occurence: int = -1

print(f'\nThe list that we will search through:\n{lst}\nThe integer we are looking for:\n{k}\n')

while True:

    if not lst:
        break
    
    list_length = len(lst)//2
    i: int = lst[list_length]
    

    if k == i:
        first_occurence = list_length + remember_index 
        lst = lst[:list_length]

    elif len(lst) == 1:
        break

    elif k < i:
        lst = lst[:list_length]
    
    else:   # k > i
        lst = lst[list_length:]
        remember_index += list_length
if first_occurence != -1:
    print(f'The first occurence of {k} was found at index {first_occurence}.\n')
  
else:
      print(f'{k} was not found in this list.\n')
```
