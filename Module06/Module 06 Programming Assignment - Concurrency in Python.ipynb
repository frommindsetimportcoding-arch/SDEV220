## Exercise 15.1
> Use *multiprocessing* to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

```python
import multiprocessing as magnus
import time as bobby
from jupyter_made_me_do_it import tha_loop


if __name__ == '__main__':
    # Increased the delay from printing so you can see that the queue was filled up in the first second or two and the time that the processes completed was much quicker than the 8 seconds that was delayed in __main__. 
    rstuvwxyz = magnus.Queue()    
    stockfish = magnus.Process(target=tha_loop, args=(rstuvwxyz,))
    stockfish.start()
    
    while stockfish.is_alive() or not rstuvwxyz.empty():
        if not rstuvwxyz.empty():
            print(f'{rstuvwxyz.get()}')
        bobby.sleep(2)
```

We were tasked with using multiprocessing and I read that using a queue is the best approach while sticking with the standard library.
